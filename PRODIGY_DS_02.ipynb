```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df1 = pd.read_csv("Titanic.csv")
df1
```

```python
df1 = df1.drop('Cabin', axis=1)
```

```python
df1.isna().sum()
```

```python
df1.isna().mean()
```

```python
df1 = df1.dropna(subset=['Embarked'])
```

```python
median_age = df1['Age'].median()
df1['Age'] = df1['Age'].fillna(median_age)
```

```python
df1.isna().sum()
```

```python
df1['Age'] = df1['Age'].round()

df1['Survived'] = df1['Survived'].map({0: 'Not Survived', 1: 'Survived'})
df1['Pclass'] = df1['Pclass'].map({1: 'Upper', 2: 'Middle', 3: 'Lower'})
df1['Embarked'] = df1['Embarked'].map({'C': 'Cherbourg', 'Q': 'Queenstown', 'S': 'Southampton'})

df1['Sex'] = df1['Sex'].str.title()

categorical_columns = ['Sex', 'Parch', 'SibSp', 'Pclass', 'Embarked', 'Survived']

df1[categorical_columns] = df1[categorical_columns].astype('category')

print(df1.info())
print(df1.head())
```

# Performing EDA

```python
# Calculate overall survival rate
survival_rate = df1['Survived'].value_counts(normalize=True)
print(f"Overall Survival Rate: {survival_rate['Survived']:.2f}")
```

```python
# Count the number of survivors and non-survivors
survival_counts = df1['Survived'].value_counts()

# Plotting the survival rate
plt.figure(figsize=(8, 6))
sns.barplot(x=survival_counts.index, y=survival_counts.values, palette='viridis')
plt.title('Overall Survival Rate')
plt.xlabel('Survival Status')
plt.ylabel('Number of Passengers')
plt.xticks(ticks=[0, 1], labels=['Not Survived', 'Survived'])
plt.ylim(0, max(survival_counts.values) * 1.1)  # Add some space above the bars

# Annotate the bars with the survival rate percentage
for index, value in enumerate(survival_counts):
    plt.text(index, value + 10, f'{(value / len(df1) * 100):.2f}%', ha='center')

plt.show()
```

```python
# Create histograms for numerical variables
ax = df1[['Age', 'Fare', 'SibSp', 'Parch']].hist(bins=20, figsize=(10, 8))

# Iterate through the axes to set labels
for axis in ax.flatten():
    axis.set_xlabel(axis.get_title(), fontsize=14)  # Set x-axis label as the title of the histogram
    axis.set_ylabel('Frequency', fontsize=14)  # Set y-axis label

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

```python
# Bar charts for categorical variables
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

sns.countplot(x='Pclass', data=df1, ax=axes[0])
sns.countplot(x='Sex', data=df1, ax=axes[1])
sns.countplot(x='Embarked', data=df1, ax=axes[2])

plt.show()
```

```python
# Calculate survival rate by Pclass
survival_by_pclass = df1.groupby('Pclass')['Survived'].value_counts(normalize=True).unstack()['Survived']
print(survival_by_pclass)

# Calculate survival rate by Sex
survival_by_sex = df1.groupby('Sex')['Survived'].value_counts(normalize=True).unstack()['Survived']
print(survival_by_sex)

# Calculate survival rate by Embarked
survival_by_embarked = df1.groupby('Embarked')['Survived'].value_counts(normalize=True).unstack()['Survived']
print(survival_by_embarked)
```

```python
# Create a figure with three subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

# Plot survival rate by Pclass
sns.barplot(ax=axes[0], x=survival_by_pclass.index, y=survival_by_pclass.values, palette='viridis')
axes[0].set_title('Survival Rate by Pclass')
axes[0].set_xlabel('Pclass')
axes[0].set_ylabel('Survival Rate')
axes[0].set_ylim(0, 1)
for index, value in enumerate(survival_by_pclass.values):
    axes[0].text(index, value + 0.02, f'{(value * 100):.2f}%', ha='center')

# Plot survival rate by Sex
sns.barplot(ax=axes[1], x=survival_by_sex.index, y=survival_by_sex.values, palette='viridis')
axes[1].set_title('Survival Rate by Sex')
axes[1].set_xlabel('Sex')
for index, value in enumerate(survival_by_sex.values):
    axes[1].text(index, value + 0.02, f'{(value * 100):.2f}%', ha='center')

# Plot survival rate by Embarked
sns.barplot(ax=axes[2], x=survival_by_embarked.index, y=survival_by_embarked.values, palette='viridis')
axes[2].set_title('Survival Rate by Embarked')
axes[2].set_xlabel('Embarked')
for index, value in enumerate(survival_by_embarked.values):
    axes[2].text(index, value + 0.02, f'{(value * 100):.2f}%', ha='center')

# Show the plot
plt.tight_layout()
plt.show()
```
